```python
def Hour(x):
    return x.split(',')[0]
```

```python
data['Time'] = pd.to_datetime(data['Time'], format = '%H:%M:%S')
```

```python
data['Hour'] = pd.to_datetime(data['Time'], format = '%H').dt.hour
```

```python
data.head()
```

```python
keys = []
hours = []
for key, hour_df in data.groupby('Hour'):
    keys.append(key)
    hours.append(len(hour_df))
```

```python
keys
hours
plt.plot(keys, hours)
plt.xlabel('Hours')
plt.ylabel('COGS')
plt.grid()
```

# There is an increase in sales between 17:00 hour to 20:00 hour of the day
